```python
import matplotlib.pyplot as plt
#print(photo)
photo = np.random.randint(low = 0, high = 255, size = (512, 512, 3))
plt.imshow(photo)
```

```python
plt.imshow(photo[::-1]) #reversed rows
```

```python
plt.imshow(photo[:, ::-1]) #mirroring , reversed colums
```

```python
plt.imshow(photo[250:750, 450:800]) #take section of the picture
```

```python
plt.imshow(photo[::2, ::2])  #take every other row & every other column
```

```python
# different math functions
photo_sin = np.sin(photo) #sine

print(np.sum(photo))
print(np.prod(photo))
print(np.mean(photo))
print(np.std(photo))
print(np.var(photo))
print(np.min(photo))
print(np.max(photo))
print(np.argmin(photo))
print(np.argmax(photo))
```
