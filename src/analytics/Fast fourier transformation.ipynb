```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft, fftfreq, ifft

## Setup for domain
# number of points
n = 1000

# Distance (meters) or time period (in secs)
Lx = 100

# Angular frequency
omg = 2.0*np.pi/Lx
omg
```

```python
## Creating individual signals
x = np.linspace(0, Lx, n)
y1 = 1.0*np.cos( 5.0*omg*x)
y2 = 1.0*np.sin(10.0*omg*x)
y3 = 0.5*np.sin(20.0*omg*x)

## Full signal
y = y1 + y2 + y3
```

```python
# Preparation
# Create all necessary frequencies
freqs = fftfreq(n)

# mask array to be used for power spectra
# ignoring half values (negative) as they are complex conjucates of the other
mask = freqs > 0

## FFT and power spectra calculations
# fft values
fft_vals = fft(y)

# true theoretical fft
fft_theo = 2.0*np.abs(fft_vals/n)

plt.figure(1)
plt.title('Original signal')
plt.plot(x, y, color='pink', label='original')
plt.legend()

plt.figure(2)
plt.title('Raw FFT values')
plt.plot(freqs[mask], fft_theo[mask], label='true fft values')
plt.show()
```
