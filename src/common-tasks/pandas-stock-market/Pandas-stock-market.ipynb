```python
import quandl
import pandas as pd
quandl.ApiConfig.api_key=''
ms = quandl.get('WIKI/MSFT')
apple = quandl.get('WIKI/AAPL')
```

```python
ms.loc['2018-01-01':'2018-03-31']['Adj. Close'].plot()
```

```python
ms_price = ms[['Adj. Close']]
apple_price = apple[['Adj. Close']]
ms_price.rename(columns={'Adj. Close':'MSFT'}, inplace=True)
apple_price.rename(columns={'Adj. Close':'AAPL'}, inplace=True)
```

```python
both_stocks = ms_price.join(apple_price,how='inner')
```

```python
both_stocks.plot()
```

```python
both_stocks.loc['2018'].plot()
```

```python
both_stocks.loc['2018'].rolling(min_periods = 10, window = 60, center = False).mean().plot()
```

```python
#How volatile is one stock vs the other
both_stocks.loc['2018'].rolling(min_periods = 10, window = 60, center = False).std().plot()
```
